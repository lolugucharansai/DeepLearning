# Logistic regression derivatives

```python
def gradient_descent(w, x, b, y, p):
    for i in range(1000):
        m=x.size
        z = np.dot(w.T, x) + b
        a = 1 / (1 + np.exp(-z))
        j = -1/m * np.sum(y * np.log(a) + (1 - y) * np.log(1 - a))
        dz = a - y
        dw = 1/m * np.dot(x, dz.T)
        db = 1/m * np.sum(dz)
        w -= p * dw
        b -= p * db
        print(j, dw, db)
```

```python

gradient_descent(w,x,1,y,0.5)
```

# Broadcasting


```python
A=np.array([[1,2,3,4],
            [3,4,5,6],
            [9,9,8,7]])

cal = A.sum(axis=0)
print(cal)
percent=100*A/cal #broadcasting
print(percent)
```
